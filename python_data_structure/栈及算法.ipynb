```python
class Stack:
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        return self.items.pop()

    def peek(self):
        return self.items[-1]

    def size(self):
        return len(self.items)
    
    def __str__(self):
        return str(self.items)
```

```python
#进制转换
def divideBy_num(head,num):
    stack = Stack()
    if head < num:
        return str(num)
    else:
        flag = True
        headnum = head
        while flag:#商用来比较
            if headnum//num < num:
                flag = False
            stack.push(headnum%num)#余数用来压栈
            headnum =  headnum//num 
        stack.push(headnum)
        return stack
#print(divideBy_num(100,8))
            
```
