```python
#冒泡排序
def bubble(alist):
    for i in range(len(alist)):
        flag = 0
        for j in range(0,len(alist)-i-1):
            if alist[j]>alist[j+1]:
                alist[j],alist[j+1] = (alist[j+1],alist[j])
                flag +=1
        if flag == 0:
            break
bubble(alist1)
alist1
```

```python
#选择排序：
def choosesort(alist):
    for i in range(len(alist)-1):
        for j in range(i,len(alist)):
            if alist[i] > alist[j]:
                alist[i],alist[j] = (alist[j],alist[i])
choosesort(alist1)
alist1
```

```python
#希尔排序  不稳定的排序 主要看参数选取
#先定义gapinsert 完成组内插入排序
  
from math import ceil
def gapinsert(alist,gap):
    for j in range(ceil(len(alist)/gap)): #组数
        start = j
        for i in range(start+gap ,len(alist),gap):#每组
            current = alist[i]
            position = i
            while alist[position - gap] > current and position >= gap:#这个条件很厉害 
                alist[position ] = alist[position-gap]
                position -= gap
            alist[position] = current
                    
def shellsort(alist):
    subgap = len(alist)//3
    while subgap > 0:
        gapinsert(alist,subgap)
        subgap  = subgap//2
shellsort(alist2)
alist2
```

```python
#快速排序 速排序不是一种稳定的排序算法
import random
def quicksort(alist):
    length = len(alist)

    first = 0
    last = length-1
    if first >= last:
        return None
    else:
        key = alist[first]
        while first<last:
            while first < last and alist[last] >= key:#第一次循环first会被取代
                last -= 1
            while first < last and alist[first] <= key: #等于号则key不动
                first +=1
            alist[first] , alist[last] = alist[last] , alist[first]
        alist[first] = key
        quicksort(alist[:first])
        quicksort(alist[first:])
```

```python
alist2 = [54,26,93,17,77,31,44,55,20] 
quicksort(alist2)
```
